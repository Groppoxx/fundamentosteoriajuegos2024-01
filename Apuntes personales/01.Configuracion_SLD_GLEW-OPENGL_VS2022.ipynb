## 4. Configurando las propiedades

* Dentro de nuestro proyecto

* Ingresamos a **Properties** en el apartado de **Project**

* Damos click en el apartado de **VC++ Directories**

* En **General** damos damos click en **Include Directories** y pegamos:

<pre> $(SolutionDir)depts/include/;$(IncludePath) </pre>

* En **General** damos damos click en **Library Directories** y pegamos:

<pre> $(SolutionDir)depts/lib/;$(LibraryPath) </pre>

* Damos click en el apartado de **Linker** y luego a **Input**

* En **Additional Dependencies** damos click y pegamos:

<pre> glew32.lib;glew32s.lib;SDL2.lib;SDL2main.lib;SDL2test.lib;opengl32.lib;%(AdditionalDependencies) </pre>

* Ejecutamos el proyecto con el siguiente codigo:

#include < SDL/SDL.h >
